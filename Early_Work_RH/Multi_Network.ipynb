```python
import numpy as np
import torch
```

```python
def preproccesing(data_x, data_y):
    #WINDOW_SIZE = 4
    #WINDOW_STEP = 1
    #NUM_PREDICT_AHEAD = 2

    pre_x = np.empty((994, 4, 30))
    pre_y = np.empty((994, 1))
    for i in range(0,994):
        for j in range(0,4):
            pre_x[i,j] = data_x[i+j]
    
    for i in range(0,994):
        pre_y[i] = data_y[i+6]

    return torch.tensor(pre_x), torch.tensor(pre_y)

    
```

```python
data_x = np.ones((1000,30))
data_y = np.ones((1000,1))

print(data_x[0:5])
print(data_y[0:5])
```

```python
pre_x, pre_y = preproccesing(data_x, data_y)

print(pre_x[0:5])
print(pre_y[0:5])
print(f"Preproccessed Data Vector X shape: {pre_x.shape}")
print(f"Preproccessed Data Vector Y shape: {pre_y.shape}")
```
